```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8, 8)})
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
time = data.datetime
count = data.count
sns.tsplot(count, time=time)
```

```python
sns.jointplot(data.temp, data.casual)
```

```python
sns.jointplot(data.windspeed, data.casual)
```

```python
c1, c2, c3 = sns.color_palette("Set1", 3)

sns.kdeplot(data.casual, shade=True, color=c1, label="casual")
sns.kdeplot(data.registered, shade=True, color=c2, label="registered")
```

```python
sns.lmplot("weather", "count", data, x_jitter=.15)
```

```python
sns.corrplot(data)
```
